# Exploring Existing LLM's Biotech Stock Analysis Capabilities

## Getting Started

In order to prepare for the development of our BioEquity Insight system, we conducted initial testing with existing LLMs to evaluate how effectively they analyze biotech stocks. This exploration helped us identify gaps in current AI capabilities and informed the design of our specialized biotech investment assistant.

**Long-Term Impact:**
- Biogen eventually announced a significant price cut (about 50%) for Aduhelm in December 2021 due to slow adoption

This FDA approval demonstrates how regulatory decisions can create both immediate price surges and subsequent corrections based on commercial realities in the biotech sector.

## Conclusions from Our Exploration

Our testing revealed several key insights:

1. **Generalist LLMs provide variable quality analysis** - While they can offer basic information about biotech stocks, they often lack detailed financial metrics, current data, or sector-specific context.

2. **Technical analysis is often superficial** - Most LLMs provide generic technical indicators without visualization or actionable insights.

3. **Specialized data + LLM analysis is superior** - Combining financial APIs with AI analysis produces the most valuable insights for investors.

4. **Biotech-specific context is lacking** - Standard LLMs often miss the nuances of regulatory impacts, clinical trial results, and pipeline valuations critical to biotech investing.

5. **Visualization capabilities are essential** - None of the tested solutions provided effective data visualization, which is critical for understanding biotech stock trends.

These findings have directly informed our development of BioEquity Insight, which addresses these gaps through specialized biotech prompting, integrated real-time financial data, customized visualization, and deep analysis of sector-specific catalysts.
